```python
import numpy as np
import random

def perceptron(x,y):
  x=np.insert(x,0,1,axis=1)
  weights=np.ones(x.shape[1])
  learning_rate=0.01

  for i in range(1000):
    j=random.randint(0, x.shape[0]-1)
    y_pred=step(np.dot(x[j],weights))
    weights=weights+learning_rate*(y[j]-y_pred)*x[j]

  return weights[0],weights[1:]
```

```python
intercept,coef=perceptron(x,y)
```
